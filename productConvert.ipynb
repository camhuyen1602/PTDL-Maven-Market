```python
#Xóa cột 'Unnamed: 0'
df.drop(['Unnamed: 0'], axis = 1,inplace = True)

#Xóa các dòng có giá trị NULL
df.dropna(inplace=True)

#Xóa các dòng trùng lặp, giữ lại dòng đầu tiên trong DataFrame
df.drop_duplicates(keep='first',inplace=True)

#Thay thế trực tiếp từ "Dalas" và "Dalla" thành "Dallas"
df["Purchase Address"].replace("Dalas,|Dalla,", "Dallas,", regex=True, inplace=True)

#Thay thế trực tiếp từ "Los Angele" thành "Los Angeles"
df["Purchase Address"].replace("Los Angele,", "Los Angeles,", regex=True, inplace=True)
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DataSales - Copy.csv")

new_columns = ['Macbook Pro Laptop', 'LG Washing Machine', 'USB-C Charging Cable',
        'AA Batteries (4-pack)', 'Bose SoundSport Headphones',
       'AAA Batteries (4-pack)', 'ThinkPad Laptop',
       'Lightning Charging Cable', 'Google Phone', 'Wired Headphones',
       'Apple Airpods Headphones', 'Vareebadd Phone', '27in FHD Monitor',
       'iPhone', '20in Monitor', '34in Ultrawide Monitor',
       'Flatscreen TV', '27in 4K Gaming Monitor', 'LG Dryer']
default_value = None

start_position = 4

for col in new_columns:
    df.insert(loc=start_position, column=col, value=default_value)
    start_position += 1
```
